```python
import numpy as np
import pandas as pd
```

## Extracting necessary columns from the given test data

```python
#loading the datasets
train_ds = pd.read_csv('training_titanic_x_y_train.csv',delimiter=',',usecols=np.arange(0,11))
df = train_ds.copy()
df.tail()
```

```python
df["Age"] = df['Age'].fillna((df['Age'].mean()))
df.tail()
```

```python
#creating another column which will represent whether the patient is female (1) or male(0)
df['f_m'] = (df["Sex"]== "female")
df['f_m'] = df['f_m'].astype(int)
df.head()
```

```python
#creating another dataframe to save the required columns
train_save = pd.DataFrame()
train_save['Sex'] = df['f_m']
train_save['Age'] = df['Age']
train_save['Survived'] = df['Survived']
train_save.reset_index()
print(train_save.head())
train_save.shape
```

```python
train_save.to_csv(r'C:\Users\jmd\Documents\Coding Ninjas ML\tr_titan.csv', index = False)
```

## Repeating the same steps for the test data

```python
test_ds = pd.read_csv("test_titanic_x_test.csv", delimiter = ",")
df = test_ds.copy()
df.head()
```

```python
#filling the NaN entries with average
df['Age'] = df["Age"].fillna((df["Age"].mean()))
df.head()
```

```python
#now creating another column to represent female as 1 and male as 0
df['f_m'] = df['Sex'] == "female"
df['f_m'] = df['f_m'].astype(int)
df.head()
```

```python
#storing the required columns in a new dataframe
save_test = pd.DataFrame()
save_test["Sex"] = df["f_m"]
save_test["Age"] = df["Age"]
save_test.reset_index()
save_test.to_csv(r'C:\Users\jmd\Documents\Coding Ninjas ML\tst_titan.csv', index = False)
save_test.head()
```

## Using Logistic Regression to predict on the basis of test data

```python
from sklearn.linear_model import LogisticRegression
```

```python
train_data = np.genfromtxt("tr_titan.csv", delimiter = ",")
test_data = np.genfromtxt("tst_titan.csv", delimiter = ",")
print(train_data.shape)
print(test_data.shape)
```

```python
#splitting into train test data
x_train = train_data[1:, 0:2]
y_train = train_data[1:, 2]

x_test = test_data[1:,:]

print(x_train.shape,  " ", y_train.shape)
print(x_test.shape)
x_train
```

```python
#defining the classifier and training the data
clf = LogisticRegression(max_iter = 1000, solver = "sag")
clf.fit(x_train, y_train)
```

```python
clf.score(x_train, y_train)
```

```python
predict = clf.predict(x_test)
predict
```

```python
#saving the predictions file
np.savetxt("titanic_prediction.csv", predict, delimiter = ",")
```
